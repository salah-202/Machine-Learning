### Import the Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Wrangling 

#### Import the dataset

```python
df = pd.read_excel('default of credit card clients.xls',header=1)
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.describe()
```

```python
df.rename({'default payment next month':'DEFAULT'},axis = 1,inplace = True)
df.head()
```

```python
df.drop('ID',axis=1,inplace=True)
df.head()
```

#### Check duplicates 

```python
df.duplicated().sum()
```

#### Checking for missing values 

```python
len(df.loc[(df['EDUCATION'] == 0) | (df['MARRIAGE'] == 0)])
```

#### Select no missing rows

```python
df_no_missing = df.loc[(df['EDUCATION'] != 0) & (df['MARRIAGE'] != 0)]
```

```python
len(df_no_missing)
```

### Resample the dataset

#### 2000 rows insted of 30,000 will be apporiate for Support vector machines

```python
df_default = df_no_missing[df_no_missing['DEFAULT'] == 1]
df_no_default = df_no_missing[df_no_missing['DEFAULT'] == 0]
```

```python
from sklearn.utils import resample

df_default_downsampled = resample(df_default,
                                 replace = False,
                                 n_samples = 1000,
                                 random_state = 42)

len(df_default_downsampled)
```

```python
df_no_default_downsampled = resample(df_no_default,
                                 replace = False,
                                 n_samples = 1000,
                                 random_state = 42)

len(df_no_default_downsampled)
```

```python
df_downsampled = pd.concat([df_default_downsampled,df_no_default_downsampled])
len(df_downsampled)
```

```python
df_downsampled.head()
```

### Split the dataset

```python
X = df_downsampled.drop('DEFAULT',axis = 1).copy()
X.head()
```

```python
y = df_downsampled['DEFAULT'].copy()
y.head()
```

### One Hot-Encoding

```python
X_encoded = pd.get_dummies(X,columns=['SEX','EDUCATION','MARRIAGE','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6'])
X_encoded.head()
```

### Centering & Scailing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale

X_train,X_test,y_train,y_test = train_test_split(X_encoded,y,random_state=42)
X_train_scaled = scale(X_train)
X_test_scaled = scale(X_test)
```

### Build a preliminary Support Vector Machine

```python
from sklearn.svm import SVC

clf = SVC(random_state=42)
clf.fit(X_train_scaled,y_train)
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(clf,X_test_scaled,y_test,values_format = 'd',display_labels=["Did not Default","Defaulted"])
```

### Optmizie pramteres with Cross Validation & Grid Search CV

```python
from sklearn.model_selection import GridSearchCV

grad_parm = [{'C':[0.5,1,10,100],
             'gamma':['scale',1,0.1,0.01,0.001,0.0001],
             'kernel':['rbf']}]

optimal_parm = GridSearchCV(SVC(),
                           grad_parm,
                           cv = 5,
                           scoring='accuracy')


optimal_parm.fit(X_train_scaled,y_train)
print(optimal_parm.best_params_)
```

### Building ,Evaluting, Drawing and Interepting the final support vector machine

```python
clf = SVC(random_state=42,gamma=0.001,C=100)
clf.fit(X_train_scaled,y_train)
```

```python
plot_confusion_matrix(clf,X_test_scaled,y_test,values_format = 'd',display_labels=["Did not Default","Defaulted"])
```
